```python
# Load the dataset for customer churn prediction into a DataFrame
df=pd.read_csv('/kaggle/input/bank-customer-churn-prediction-dataset/Churn_Modelling.csv')
```

```python
from sklearn.ensemble import RandomForestClassifier  # Import classifier for churn prediction
from sklearn.model_selection import train_test_split  # Import to split data into training and test sets

df.head(3)  # Display the first 3 rows of the dataset to check initial data structure if we use df.head() just show the first 3 rows of the dataset
```

```python
df = df.drop(columns=["RowNumber", "CustomerId", "Surname"])  # Remove unnecessary columns for prediction
df.head(3)  # Display the first 3 rows to confirm column removal
```

```python
df = pd.get_dummies(df, columns=["Geography", "Gender"], drop_first=True)  # Convert categorical columns to dummy variables
```

```python
y = df["Exited"]  # Define target variable for prediction (whether customer exited)
x = df.drop("Exited", axis=1)  # Define feature set by dropping the target column & you can use x = df.drop(columns=['Exited'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=61,train_size=0.7) # Split data into training and test sets
```

```python
rf = RandomForestClassifier()  # Initialize the random forest classifier
model = rf.fit(x_train, y_train)  # Train the model on the training set
model.score(x_test, y_test)  # Evaluate model accuracy on the test set
```

```python
model.predict([[712, 31, 9, 0, 5, 1, 1, 200000, False, False, True]])  # Predict customer churn for given features
```

```python
from sklearn.preprocessing import OneHotEncoder  # Import the OneHotEncoder for encoding categorical variables
```

```python
df = pd.read_csv('/kaggle/input/bank-customer-churn-prediction-dataset/Churn_Modelling.csv')  # Load the dataset into a DataFrame
df = df.drop(columns=["RowNumber", "CustomerId", "Surname"])  # Remove unnecessary columns for analysis
```

```python
ohe = OneHotEncoder()  # Initialize the OneHotEncoder for converting categorical variables
```

```python
encoded_features = ohe.fit_transform(df[['Geography', 'Gender']]).toarray()  # Apply one-hot encoding to the specified categorical columns
```

```python
encoded_features
```

```python
ohe.get_feature_names_out()  # Retrieve the names of the one-hot encoded features
```

```python
encoded_features = pd.DataFrame(encoded_features)  # Convert the encoded features array into a DataFrame
```

```python
encoded_features.columns = ohe.get_feature_names_out()  # Set the column names of the DataFrame to the encoded feature names
```

```python
encoded_features
```

```python
df.head(3)
```

```python
df=df.drop(columns=['Geography','Gender'])
df.head(3)
```

```python
df[encoded_features.columns] = encoded_features  # Add the one-hot encoded features to the original DataFrame
df.head(3)
```

```python
y=df["Exited"]
x=df.drop("Exited",axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier  # Import the DecisionTreeClassifier for classification tasks
```

```python
tree = DecisionTreeClassifier()  # Initialize the Decision Tree classifier
model = tree.fit(x, y)  # Train the model on the feature set (x) and target variable (y)
model.score(x, y)  # Evaluate the model accuracy on the same data to check for overfitting
# overfitting
```
